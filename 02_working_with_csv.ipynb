```python
import pandas as pd
```

```python
# without separator
df = pd.read_csv('data_samples/sample_data.tsv')
display(df[:2])
df2 = pd.read_csv('data_samples/sample_data_semicolon.csv')
display(df2[:2])

# separator
df = pd.read_csv('data_samples/sample_data.tsv', sep='\t')
display(df[:2])
df2 = pd.read_csv('data_samples/sample_data_semicolon.csv', sep=';')
display(df2[:2])
```

```python
# index_col
df = pd.read_csv('data_samples/sample_data.csv', index_col='ID')
display(df[:2])
```

```python
# specify header
dfh = pd.read_csv('data_samples/sample_data.csv', header=0)
display(dfh[:2])
dfh = pd.read_csv('data_samples/sample_data.csv', header=1)
display(dfh[:2])
```

```python
# load specific cols
dfc = pd.read_csv('data_samples/sample_data.csv', usecols=['ID', 'Name', 'Age'])
display(dfc[:2])
```

```python
# squeeze, make single col a pandas series
dfs = pd.read_csv('data_samples/sample_data.csv', usecols=['Age']).squeeze()
display(dfs[:2])
display(type(dfs))

# does not work for more than 1 col
dfs = pd.read_csv('data_samples/sample_data.csv', usecols=['Age', 'Name']).squeeze()
display(dfs[:2])
display(type(dfs))
```

```python
dfk = pd.read_csv('data_samples/sample_data.csv')
display(dfk[:4])
# skiprows
dfk = pd.read_csv('data_samples/sample_data.csv', skiprows=[1,2])
display(dfk[:2])
# skiprows range
dfk = pd.read_csv('data_samples/sample_data.csv', skiprows=range(1,51))
display(dfk[:2])
```

```python
# nrows, load specified # of rows
dfn = pd.read_csv('data_samples/sample_data.csv', nrows=3)
display(dfn)
display('shape:', dfn.shape)
```

```python
# encoding
dfe = pd.read_csv('data_samples/sample_data.csv', encoding='utf-8') # latin-1, ascii, unicode
display(dfe[:3])
```

```python
# dfb = pd.read_csv('data_samples/sample_data_bad.csv') 
# # error: Error tokenizing data. C error: Expected 5 fields in line 9, saw 6

# ignore bad lines
dfb = pd.read_csv('data_samples/sample_data_bad.csv', on_bad_lines='skip')
display(dfb[:2])
```

```python
# dtype, specify each col's type before loading data
dfcl = pd.read_csv('data_samples/sample_data.csv', dtype={'ID': int, 'Name': str, 'Age': float})
dfcl.info()
```

```python
# date data parsing
dfd = pd.read_csv('data_samples/sample_data_date_parsing.csv')
display(dfd.info())

dfd2 = pd.read_csv('data_samples/sample_data_date_parsing.csv', parse_dates=['Date'])
display(dfd2.info())
```

```python
dfsal = pd.read_csv('data_samples/sample_data.csv')
display(dfsal[:2])

# converters, apply a function to cols
def bonus(salary):
    return float(salary) + 30000

dfsal = pd.read_csv('data_samples/sample_data.csv', converters={'Salary': bonus})
display(dfsal[:2])
```

```python
# na_values, handle null/missing values
dfn = pd.read_csv('data_samples/sample_data_missing_values.csv')
display(dfn[:7])

dfn2 = pd.read_csv('data_samples/sample_data_missing_values.csv', na_values=['-'])
display(dfn2[:7])
```
