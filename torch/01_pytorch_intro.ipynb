```python
import torch
import pandas as pd
import numpy as np
```

```python
# vectors
temp = torch.FloatTensor([[4,5,3,2.2],[3,7,8,3]])
print(temp)
print(temp.size())
print(temp.shape)
print(temp[1])
print(temp[1,3],'\n')

temp2 = torch.rand(3,5)
print(temp2)
print(temp2.size())
print(temp2.shape)
print(temp2[1])
print(temp2[1,1])
```

```python
# scalars
x = torch.rand(10)
print(x)
print(x.size())
print(x.shape)
```
