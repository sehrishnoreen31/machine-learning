```python
# count plot for categorical cols
plt.figure(figsize=(5,3))
colors = ['orange', 'blue', 'red', 'green', 'skyblue']
df['Sunlight_Exposure'].value_counts().plot(kind='bar', color=colors)
plt.title('Sunlight_Exposure')
plt.show()
```

```python
# pie chart for categorical cols
plt.figure(figsize=(3,5))
df['Soil_Type'].value_counts().plot(kind='pie', autopct='%.2f')
plt.xlabel(None)
plt.ylabel(None)
plt.title('Soil_Type')
```

```python
# histogram for numeric 
plt.figure(figsize=(6,3))
df['Watering_Amount_ml'].plot(kind='hist', bins=50)
plt.title('Watering_Amount_ml')
```

```python
# distplot for the numeric cols
fig, (ax1, ax2, ax3) = plt.subplots(1,3, figsize=(12,3))
df['Height_cm'].plot(kind='kde', color='blue', ax=ax1,)
ax1.set_title('Height_cm')
df['Humidity_%'].plot(kind='kde', color='green', ax=ax2)
ax2.set_title('Humidity_%')
df['Soil_Moisture_%'].plot(kind='kde', color='red', ax=ax3)
ax3.set_title('Soil_Moisture_%')
```

```python
# boxplot for numeric data, for 5 number summary and outliers
fig, (ax1, ax2, ax3) = plt.subplots(1,3, figsize=(10,3))
df['Health_Score'].plot(kind='box',color='blue', ax=ax1)
df['Watering_Frequency_days'].plot(kind='box',color='red', ax=ax2)
df['Room_Temperature_C'].plot(kind='box',color='magenta', ax=ax3)
```
